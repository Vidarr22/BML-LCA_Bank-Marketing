```python
import numpy as np
from sklearn.decomposition import PCA
import pandas as pd

# Step 1: Define data
X = np.array([
    [4, 11],
    [8, 4],
    [13, 5],
    [7, 14]
])

# Step 2: Mean-center (no scaling)
X_centered = X - X.mean(axis=0)

# Step 3: Apply PCA
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X_centered)

# Step 4: Flip sign if needed to match table direction
X_pca = -X_pca  # just flips the direction, doesnâ€™t affect variance

# Step 5: Create output DataFrame
df = pd.DataFrame({
    "X1": X[:, 0],
    "X2": X[:, 1],
    "First_Principal_Component": np.round(X_pca.flatten(), 4)
})

print(df)
print("\nExplained Variance Ratio:", np.round(pca.explained_variance_ratio_, 4))
print("Principal Component (Eigenvector):", np.round(pca.components_, 4))
```
